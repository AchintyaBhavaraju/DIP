```python
import cv2 as cv    #Importing open cv
import matplotlib.pyplot as plt
from PIL import Image as im
import numpy as np

image=cv.imread('C:/Users/achin/Downloads/cube.jpg')  #Reading image
cv.imshow('Image',image)  #Displaying all kind of images
cv.waitKey(0)

image.reshape(3066,144,4)#Image reshaping
cv.imshow('Image',image)  #Displaying all kind of images

grayscale=cv.cvtColor(image,cv.COLOR_BGR2GRAY)  #Coverting to grayscale
cv.imshow('Image',grayscale)  #Displaying all kind of images

(thresh,blackAndWhiteImage)=cv.threshold(grayscale,128,255,cv.THRESH_OTSU)
cv.imshow('Image',blackAndWhiteImage)  #Displaying all kind of images

#image profile
profile = cv.calcHist([image],[0],None,[256],[0,256])
plt.plot(profile)
plt.show()
#plots the image profile

(B,G,R)=cv.split(image)   #Splitting in RGB plane
cv.imshow('Red',R)  #Displaying all kind of images
cv.imshow('Green',G)
cv.imshow('Blue',B)

merged=cv.merge([B, G, R])# Color image using R,G and B planes
cv.imshow('Blue',merged)

#flow control and loop in Python
count=0
for i in range(0,10):
    count=count+1
if count>5:
    print("Greater than 5")
else:
    print("Not greater than 5")

newarray = np.arange(0, 625000, 1, np.uint8)
newarray = np.reshape(newarray, (1250, 500))
print(newarray)
new_img = im.fromarray(newarray)
new_img.save('new.png') 
```
